```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
```

```python
import ast

def convert_cast_x(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3 :
            l.append(i['name'])
        counter += 1

    return l
```

```python
movies['cast_x'] = movies['cast_x'].apply(convert_cast_x)
movies.head(2)
```

```python
import ast

def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
             l.append(i['name'])
             break
 

    return l
```

```python
movies['crew_x'] = movies['crew_x'].apply(fetch_director)
movies.head(2)
```

```python
movies.iloc[0]['overview']
```

```python
movies.head()
```

```python
def remove_space(word):
    l =[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast_x'] = movies['cast_x'].apply(remove_space)
movies['crew_x'] = movies['crew_x'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

movies.head()

```python
movies.head()
```

```python
movies['tags']  =  movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast_x'] + movies['crew_x']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id_x','title','tags']]
```

```python
new_df = movies[['movie_id_x','title','tags']].copy()
```

```python
new_df.head()
```

```python
new_df['tags'] =  new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df['tags'][2]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))

        return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```
